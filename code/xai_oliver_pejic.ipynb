## Setup <a id='toc2_'></a>

```python
import torch
import numpy as np
import torch.nn.functional as F
import torch.nn as nn
import matplotlib.pyplot as plt
import warnings # to suppress warnings
warnings.filterwarnings("ignore") # ignore warnings
```

## Load and Define the CNN Model <a id='toc3_'></a>

Here we define the CNN model that will be used to classify the digit images.  
This model is simple and works well for MNIST-like data.

- It has two convolutional layers (`conv1` and `conv2`) to extract features from the image.  
- After each convolution, we use ReLU activation and MaxPooling to reduce size and keep important features.  
- Then we flatten the image and pass it through two fully connected layers to make the final prediction.  
- The final output layer has 10 units—one for each digit (0 to 9).

After defining the model, we load its best saved weights and set it to evaluation mode to make predictions.

```python
# Define the CNN model
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64 * 7 * 7)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
    
# Load the models best weights
model = SimpleCNN()
model.load_state_dict(torch.load('../models/challenge_model.pth'))
model.eval()
```

## Vanilla Saliency Maps <a id='toc7_'></a>

We define a function to create **Vanilla Saliency Maps**, one of the simplest XAI methods.

- It works by computing the gradient of the model’s output with respect to the input image.  
- This shows which pixels had the strongest influence on the prediction.  
- The result is a heatmap where bright areas mean "important pixels" for the decision.

We apply it to the selected image. The result is a 28×28 map, showing the same size as the original image.

```python
def vanilla_saliency(model, input_tensor, target):
    x = input_tensor
    
    # Ensure input is a 4D tensor (N, C, H, W)
    if x.dim() == 5:
        x = x.view(-1, 1, x.size(-2), x.size(-1))
    if x.dim() == 3:
        x = x.unsqueeze(1)
    if x.dim() == 2:
        x = x.unsqueeze(0).unsqueeze(0)

    # Ensure the input tensor is on the same device as the model
    x = x.clone().detach().requires_grad_(True)
    out = model(x)[0, target]
    out.backward()

    # Get the absolute value of the gradient
    sal = x.grad.abs().squeeze().cpu().numpy()
    return sal
```

## Integrated Gradients <a id='toc8_'></a>

Next, we use **Integrated Gradients**, a more advanced XAI method.

- It compares the input image to a baseline (a black image) and measures how the model’s prediction changes as we move from the baseline to the real image.  
- This method gives smoother and more reliable explanations than plain gradients.

We apply Integrated Gradients to the first image in the dataset.  
The result is again a 28×28 heatmap showing which areas were important for the prediction.

```python
def integrated_gradients(model, input_tensor, target, baseline=None, steps=50):

    x0 = input_tensor

    # Ensure input is a 4D tensor (N, C, H, W)
    if x0.dim() == 5:
        x0 = x0.view(-1,1,x0.size(-2),x0.size(-1))
    if x0.dim() == 3:
        x0 = x0.unsqueeze(1)
    if x0.dim() == 2:
        x0 = x0.unsqueeze(0).unsqueeze(0)
    if baseline is None:
        baseline = torch.zeros_like(x0)

    scaled = [baseline + (i/steps)*(x0-baseline) for i in range(steps+1)]
    grads = []

    for img in scaled:
        img = img.clone().detach().requires_grad_(True)
        out = model(img)[0, target]
        out.backward()
        grads.append(img.grad.detach().cpu().numpy())

    grads = np.array(grads)
    avg_grads = (grads[:-1] + grads[1:]) / 2  
    delta = (x0 - baseline).cpu().detach().numpy()
    ig = delta * avg_grads.mean(axis=0)    

    return ig.squeeze() 
```

```python
# Example usage of integrated gradients
img = torch.tensor(challenge_images[0], dtype=torch.float32).unsqueeze(0).unsqueeze(0)
ig = integrated_gradients(model, img, int(challenge_labels[0]))
print(ig.shape)
```

## Occlusion <a id='toc9_'></a>

Now we apply the **Occlusion** method to explain the model’s decision.

- This method works by sliding a small gray patch (e.g., 4×4) over the image and checking how the prediction changes.  
- If covering a region causes the model to become less confident, that region is important.  
- The result is a heatmap showing which parts of the image matter most.

We apply occlusion to the first image using a patch size of 4×4.  
The output is a 28×28 map showing the impact of each region.

```python
from captum.attr import Occlusion

occlusion = Occlusion(model)


def occlusion_heatmap(model, input_tensor, target,
                      window_shape=(1,4,4), stride=(1,4,4)):
    
    x = input_tensor
    if x.dim() == 5: 
        x = x.view(-1, 1, x.size(-2), x.size(-1))
    if x.dim() == 3:
        x = x.unsqueeze(1)
    if x.dim() == 2:
        x = x.unsqueeze(0).unsqueeze(0)

    attributions = occlusion.attribute(
        inputs=x,
        baselines=torch.zeros_like(x),
        target=target,
        sliding_window_shapes=window_shape,
        strides=stride
    )

    heatmap = attributions.mean(dim=1).squeeze(0).cpu().numpy()
    return heatmap
```

```python
# Example usage of occlusion
img = torch.tensor(challenge_images[0], dtype=torch.float32).unsqueeze(0).unsqueeze(0)
heatmap = occlusion_heatmap(model, img, int(challenge_labels[0]), window_shape=(1, 4, 4), stride=(1, 4, 4))
print(heatmap.shape)
```

## Grad-CAM <a id='toc10_'></a>

Finally, we use **Grad-CAM (Gradient-weighted Class Activation Mapping)**.

- This method uses the gradients of the last convolutional layer to find important regions in the image.  
- It creates a coarse heatmap showing which parts of the image activated deeper layers in the model.  
- Grad-CAM is especially useful for understanding what the model sees at a higher, more abstract level.

We apply Grad-CAM to the first image using the second convolutional layer (`conv2`).  
The result is a 28×28 heatmap that highlights general regions the model focused on when making its decision.

```python
from captum.attr import LayerGradCam

target_layer = model.conv2
lgc = LayerGradCam(model, target_layer)
def gradcam_heatmap(model, input_tensor, target, layergc):

    x = input_tensor
    if x.dim() == 5:
        x = x.view(-1, 1, x.size(-2), x.size(-1))
    if x.dim() == 3:
        x = x.unsqueeze(1)
    if x.dim() == 2:
        x = x.unsqueeze(0).unsqueeze(0)

    attributions = layergc.attribute(inputs=x, target=target)

    upsampled = torch.nn.functional.interpolate(
        attributions,
        size=(x.shape[-2], x.shape[-1]),
        mode='bilinear',
        align_corners=False
    )

    sal = upsampled.squeeze(0).sum(dim=0).cpu().detach().numpy()
    sal = (sal - sal.min()) / (sal.max() - sal.min() + 1e-8)
    
    return sal
```

```python
# Example usage of gradcam
img = torch.tensor(challenge_images[0], dtype=torch.float32).unsqueeze(0).unsqueeze(0)
heatmap = gradcam_heatmap(model, img, int(challenge_labels[0]), lgc)
print(heatmap.shape)
```

## Visual Comparison of All XAI Methods <a id='toc11_'></a>

We now apply all four XAI methods to each image in the challenge set and display them side by side.

- For every image, we show:
  - The original digit image  
  - Saliency map  
  - Integrated Gradients  
  - Occlusion  
  - Grad-CAM

This visual comparison helps us understand how each method highlights different parts of the image.  
It also shows which explanations are clear, which are noisy, and how well the model's focus matches the digit shape.

```python
methods = [
  ("Original",           lambda x,y: x.squeeze()),
  ("Saliency",           lambda x,y: vanilla_saliency(model, x, y)),
  ("IntegratedGradients",lambda x,y: integrated_gradients(model, x, y)),
  ("Occlusion",          lambda x,y: occlusion_heatmap(model, x, y)),
  ("GradCAM",            lambda x,y: gradcam_heatmap(model, x, y, lgc)),
]

for i in range(len(challenge_images)):
    img = challenge_images[i]
    label = int(challenge_labels[i])
    inp = torch.tensor(img, dtype=torch.float32).unsqueeze(0).unsqueeze(0)

    fig, axes = plt.subplots(1, 5, figsize=(10,2))
    for ax, (title, fn) in zip(axes, methods):
        heat = fn(inp, label)
        ax.imshow(img.squeeze(), cmap='gray' if title=="Original" else None)
        if title!="Original":
            ax.imshow(heat, cmap='hot', alpha=0.4)
        ax.set_title(title)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

## Cross‐Method Comparison <a id='toc13_'></a>

| Method                | Main Focus                         | Key Advantage                                | Key Disadvantage                         |
|-----------------------|------------------------------------|-----------------------------------------------|------------------------------------------|
| **Vanilla Saliency**  | Strong pixels (high gradient)       | Very fast and easy to use                     | Often noisy, sometimes highlights wrong or random spots |
| **Integrated Gradients** | Whole shapes and smooth lines         | Gives clear and smooth results                | Slower to compute (many steps needed)     |
| **Occlusion**         | Important image parts by masking     | Easy to understand—tests what happens when parts are hidden | Can look blocky and depends on window size |
| **Grad-CAM**          | Big regions in deeper layers         | Shows where the model looks in general        | Low detail and sometimes blurry          |


## Conclusions & Recommendations <a id='toc14_'></a>

- **When to use which method?**  
  - Use **Vanilla Saliency** to quickly check what the model is focusing on.  
  - Use **Integrated Gradients** when you want a clear and full explanation of the digit’s shape.  
  - Use **Occlusion** to test which areas are most important by hiding them.  
  - Use **Grad-CAM** to see what parts of the image the model uses overall.

- **Best method for explaining handwritten digits:**  
  **Integrated Gradients** shows the most complete and smooth shapes.  
  **Occlusion** helps to confirm important parts by hiding them.

- **Things to be careful about:**  
  - **Saliency** can be noisy and hard to read.  
  - **Occlusion** can look rough if the patch size is too big.  
  - **Grad-CAM** can miss small details like thin lines.

## Which XAI Methods Work Best for MNIST? <a id='toc15_'></a>

The MNIST dataset contains small, simple black-and-white digit images. Good explanations need to show **clear shapes**, **strokes**, and **important areas** of the digit.

Here’s what works well:

- **Integrated Gradients** is very effective.  
  It gives smooth and full outlines of digits like “3”, “7”, or “8”. The shapes are easy to understand and not noisy.

- **Occlusion** also works well.  
  It clearly shows which regions matter by testing what happens when parts of the image are hidden. This gives useful feedback, especially on curves or crossings.

- **Vanilla Saliency** can work for digits with simple shapes (like “1” or “7”).  
  It highlights important pixels quickly, but is less helpful for complex shapes like “8” or “3”, where it becomes noisy.

- **Grad-CAM** gives a general view of where the model looks.  
  It works better on larger images with more structure. For MNIST, it often misses small details, like thin strokes.

**Conclusion:**  
For MNIST, use **Integrated Gradients** and **Occlusion** for best results.  
**Saliency** is fast and works okay for simple digits, but not for all.  
**Grad-CAM** is more general and less useful for fine details in this dataset.
